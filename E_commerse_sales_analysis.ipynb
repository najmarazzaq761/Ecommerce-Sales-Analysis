## Highest Price Product

```python
highest_price_product=df.loc[df["price"].idxmax(),["category","price"]]
print(highest_price_product)

```

```python
highest_price_product=df.loc[df["price"].idxmax()]
print(highest_price_product)
```

## Lowest Price Product

```python
lowest_price_product=df.loc[df["price"].idxmin(),["product_name","price"]]
print(lowest_price_product)
```

## Which product has highest number of reviews?

```python
highest_review_product=df.loc[df["review_count"].idxmax(),["product_name","review_count"]]
print(highest_review_product)
```

## total review counts by category/ Which category has highest number of reviews?

```python
category_reviews=df.groupby("category")["review_count"].sum().reset_index()
# plotting
plt.figure(figsize=(10,6))
sns.barplot(x="category",y="review_count",data=category_reviews)
plt.xlabel("Category")
plt.ylabel("Total Reviews")
plt.title("Total Reviews counts by Category")
plt.show()
```

```python
df.head()
```

## count review_score (rating)  greater then 4 for each category and show which has more

```python
high_rating=df[df["review_score"]>4]
category_high_rating_counts=high_rating.groupby("category")["review_score"].count().reset_index(name="counts")
# plotting
plt.figure(figsize=(10,6))
sns.barplot(x="category",y="counts",data=category_high_rating_counts,color="green")
plt.xlabel("categories")
plt.ylabel("Count")
plt.title("Number of high ratings(greater then 4) by Category")
plt.show()
```

## Show the sales graph for any specified category for the entire year

```python
specific_category = 'Sports'

# Grouping by category and summing sales for each month
grouped_df = df2.groupby('category').sum()

# # Check if the specified category exists
# if specific_category in grouped_df.index:
#     # Extract the data for the specified category
category_sales = grouped_df.loc[specific_category]
# Plotting
plt.figure(figsize=(14, 6))
plt.plot(category_sales.index, category_sales.values, marker='o')
plt.title(f'Sales Analysis for {specific_category} (12 Months)')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.legend(title='Category')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
# else:
#     print(f"Category '{specific_category}' not found in the data.")
```

## Show which category of product has more sales over  the entire year

```python
grouped_df = df2.groupby('category').sum()

# Summing total sales for the year for each category
grouped_df['Total Sales'] = grouped_df.sum(axis=1)

# Plotting total sales for each category
plt.figure(figsize=(10, 6))
plt.bar(grouped_df.index, grouped_df['Total Sales'], color='skyblue')
plt.title('Total Sales by Category (Full Year)')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

## heatmap

```python
# Filter the DataFrame to include only numeric columns
numeric_df = df.select_dtypes(include=[float, int])

# Calculate the correlation matrix
correlation = numeric_df.corr()

# Plot the heatmap
plt.figure(figsize=(14, 8))
sns.heatmap(correlation, cmap="BrBG", annot=True)
plt.show()
```
